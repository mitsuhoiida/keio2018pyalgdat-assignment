```python
def check(key,other):
    if len(key) > 2:
        if len(other) > 2:
            if ((key[0]==other[0]) and (key[1]==other[1]) and (key[2]==other[2])):
                return 1
            else:
                return 0
        else:
            return 0
    else:
        return 0            
```

```python
check('nonplused', data['nonplused'][0][0])
```

```python
def listup(inp):
    output = []
    for x in inp:
        num = check(x,inp[x][0][0])
        if num==1:
            output.append(x)
    return output
```

```python
remove = listup(data)
```
